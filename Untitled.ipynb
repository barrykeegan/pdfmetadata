```python
# want to search for first forward slash b'^/' followed by alphabet (alpha num might do '\w') characters 
# followed by up to word boundary (non alpha num should do this too '\W')
exp = re.compile(b'/\w*')
```

```python
pdf_obj_dict = {}
for line in test_list_no_FlateDecode:
    m = exp.match(line)
    print(f'm={m},\nm.group()={m.group()},\nm.groups()={m.groups()}\n\n')
    if m.group() in pdf_obj_dict.keys():
        pdf_obj_dict[m.group()].append(line)
    else:
        pdf_obj_dict[m.group()] = [line]
        
    #print(pdf_obj_dict[m.group()])
print(f'pdf_obj_dict.keys()={pdf_obj_dict.keys()}')
```

```python
#excluded_objects = []
excluded_objects = [b'/Linearized']
for key in pdf_obj_dict.keys():
    if key in excluded_objects:
        continue
    print(f"Object Name: {key}")
    for line in pdf_obj_dict[key]:
        print(f"{line}\n")
    print()
```
